# Decision Tree Regressor

```python
from decision_trees import decision_tree_regressor
dt_methods = decision_tree_regressor()
```

Decision trees return the mean absolute error, prints the score like the original notebook, and then plots the tree (unless you set that parameter to false).

```python
#default method decides num nodes 
dt_methods.auto_decision_tree(my_dataframe.iloc[:, 2:6], my_dataframe['nbo_N_oxamide_max'],  plot=True)
```

You can also use any additional sklearn methods you'd like on the tree, such as max_depth

https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier

```python
dt_methods.tree.max_depth
```

```python
#you can determine the number of nodes >1 
dt_methods.decision_tree(my_dataframe.iloc[:, 2:6], my_dataframe['nbo_N_oxamide_max'], num_nodes = 5, plot=True)
```

```python
#or you can try a single node (like the old matlab notebook)
dt_methods.single_node_decision_tree(my_dataframe.iloc[:, 2:6], my_dataframe['nbo_N_oxamide_max'], plot=False)
```

The results of the most recent tree are saved as a decision tree object until a new tree is generated. 

To delete that tree:

```python
dt_methods.delete_tree()
```
