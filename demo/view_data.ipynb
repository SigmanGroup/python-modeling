# Dynamic File Reader Options

```python
'''
Automate construction of dataframe (rather than have the user provide certain params)
'''

def read_excel_file(path=0, sheet=0):
    #read file and create pandas DataFrame
    try:
        df = pd.read_excel(io=path, sheet_name=sheet, header=5)
        df.dropna(axis = 0, how = 'all', inplace = True)
        return df
    #error message if file cannot be found
    except IOError:
        print ("Could not read file:", path)
```

```python
read_excel_file("../model_oct_bothsubs"+".xlsx")
```
